# Remove all genes that are not 1:1 orthologs

```python
# read in the list of elegans genes and and their orthologues (if any) in briggsae
orthologous_genes=pd.read_csv('/mnt/c/Users/helen/hrh/part_ii_project/orthologs/elegans_briggsae_with_ortholog_type.txt', sep='\t')
orthologous_genes[:3]
```

```python
# take only genes that are 1:1 orthologs
orthologs_of_interest=orthologous_genes[orthologous_genes['Homology type']=='ortholog_one2one'].copy().drop('Homology type', axis=1)
orthologs_of_interest[:3]
```

```python
len(orthologs_of_interest)
```

```python
# make a dictionary to get from elegans gene to the c briggsae ortholog (not directly relevant, but handy)
elegans_to_briggsae={}

for index, row in orthologs_of_interest.iterrows():
    elegans_to_briggsae[row['Gene stable ID']]=row['Caenorhabditis briggsae (PRJNA10731) [WS276] gene stable ID']
```

```python
# dataframe same as orthologs_of_interest but with nicer column names
nicer=pd.DataFrame({'elegans':list(orthologs_of_interest['Gene stable ID']),
                   'briggsae':list(orthologs_of_interest['Caenorhabditis briggsae (PRJNA10731) [WS276] gene stable ID'])})
```

```python
len(elegans_genes_gd)
```

```python
# this goes through all the genes in the c. elegans gene list and takes only those genes where:
# 1) they are present in the list of 1:1 orthologs
# 2) their briggsae orthologue is present in the briggsae gene list (some discrepances here due to release #)

# it also gathers the indices necessary to reorder the briggsae df so that each ortholog in elegans/briggsae is at the same index

elegans_good_genes=[]
elegans_idx=[]
briggsae_good_genes=[]
briggsae_idx=[]
for index, row in elegans_genes_gd.iterrows():
    take=nicer[nicer['elegans']==row['name']].copy().reset_index(drop=True)
    if len(take)>0:
        b_take=briggsae_genes_gd[briggsae_genes_gd['name']==take['briggsae'][0]].copy().reset_index(drop=False)
        if len(b_take)>0:
            elegans_good_genes.append(row['name'])
            elegans_idx.append(index)
            briggsae_good_genes.append(b_take['name'][0])
            briggsae_idx.append(b_take['index'][0])
```
